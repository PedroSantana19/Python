```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,5, 11)
y= x * x
```

```python
plt.plot(x,y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('TÃ­tulo')
```

```python
plt.plot(x,y, color='R')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('TÃ­tulo')
```

```python
plt.plot(x,y, 'r--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('TÃ­tulo')
```

```python
plt.subplot(1, 2 , 1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
```
