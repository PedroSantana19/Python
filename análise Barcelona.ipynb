```python
#Criar novo conjunto de dados pegando todos os dados da linha onde a coluna "Club" for FC Barcelona#
ssm = df.query('Club == "FC Barcelona"')
```

```python
ssm.shape
```

```python
#Agrupando os dados por nacionalidade e vendo o numero de jogadores de cada pa√≠s#
df.groupby('Nationality').size()
```

```python
#Agrupando os dados por nacionalidade e vendo a m√©dia do portencial dos jogadores desses pa√≠ses#
df.groupby('Nationality').Potential.mean().round(1)
```

```python
df['JogadoresRuins'] = df.Potential < 50
```

```python
sns.lmplot('Age', 'Overall', hue='JogadoresRuins', data=df)
```

```python
plt.title('Potencial dos Jogadores do Barcelona')
ax = sns.countplot(x='Potential', hue='Name', data=ssm, )
plt.xlabel('Potencial')
plt.ylabel('')
ax.legend(title='Nome', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
